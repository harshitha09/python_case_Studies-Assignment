```python
import pandas as pd
import numpy as np
```

```python
terror = pd.read_csv("terror.csv", engine = "python")
terror
```

```python
list(terror)
```

```python
# 1.How many attacks happened in India?

len(terror.query("country_txt == 'India'"))
```

```python
# 2.How many attacks happened in India and upto 3 people were killed? 

len(terror[(terror.country_txt == 'India') & (terror.nkill <= 3)])
```

```python
# 3.Extract the city and summary for attacks above

terror[(terror.country_txt == 'India') & (terror.nkill <= 3)][["city","summary"]]
```

```python
# 4.In a single terror incident in India, find out top 5 cities by number killed

terror[(terror.nperps == 1) & (terror.country_txt == 'India')].sort_values('nkill',ascending = False)["city"].head(5)
```

```python
# 5.In a single terror incident in India, find out top 5 cities by number killed and wounded 


terror["kill_wound"] = terror.nkill + terror.nwound
terror[(terror.nperps == 1) & (terror.country_txt == 'India')].sort_values('kill_wound',ascending = False)["city"].head()
```

```python
# 6.How many attacks were successful that were suicide attacks?

len(terror[(terror.success == 1) & (terror.suicide == 1)])
```

```python
# 7.label all the incidents where the number killed was more than 5 as severe

terror["severe"] = np.nan
terror.loc[terror.nkill > 5, "severe"] = "severe"
print(terror.severe)
```

```python
# 8.write a function to label an incident that was both successful and suicidal

def incident():
    terror.loc[(terror.success == 1) & (terror.suicide == 1), "severe"] = "incident"
    
incident()
```

```python
# 9.Create a new category representing if the incident occured in Afghanistan, 
# Pakistan or India as one level of the category and all the other countries as another level

terror["category_new"] = np.nan
API = ["Afganisthan","Pakistan","India"]
terror.loc[terror.country_txt.isin(API), "category_new"] = "Asia" 
terror.loc[~(terror.country_txt.isin(API)),"category_new"] = "rest_countries"
terror["category_new"]
```

```python
# 10.How many incidents happened in Af-Pak-India vs ROW?

terror["category_new"].value_counts()

```

```python
# 11.List the number of suicides attacks and average kills by Af-Pak-India vs ROW. 
# Rename columns in the output as Average_Kills and Number_Incidents.

pd.crosstab(terror.category_new,terror.suicide,colnames = ["Number_Incidents"])
df = terror.groupby("category_new")["nkill"].mean().reset_index()
df.columns = ["category","Average_kills"]
df
```
